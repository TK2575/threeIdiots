```python
from pybaseball import standings
import pandas as pd
```

```python
# import regular-season-futures-selections.csv as pandas data frame
selections = pd.read_csv('regular-season-futures-selections.csv')
selections
```

```python
# collapse data into one dataframe
df = pd.concat(standings(2023)).sort_values(by=['W-L%'], ascending=False).reset_index(drop=True)
df.info()
```

```python
# join standings and df by team name
results = selections.join(df.set_index('Tm'), on='team')
# add points column to results, equal to wins if inverted is false and loss if inverted is true
results['points'] = results.apply(lambda x: x['L'] if x['inverted'] else x['W'], axis=1).astype(int)
# convert points to integer
results.sort_values(by=['points'], ascending=False).reset_index(drop=True)
```

```python
# summarize results by player and sum points sort by points descending
results.groupby(['player'], as_index=False)['points'].sum().sort_values(by=['points'], ascending=False).reset_index(drop=True)
```
